```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def streamline(r, a, b, const):
    eq = (a/b)*np.log(r)+const
    return eq
```

```python
r = np.linspace(0.01,100,1000)
```

```python
#cyl --> cart
def cart(r,theta):
    x = r*np.cos(theta)
    y = r*np.sin(theta)
    return(x,y)
```

```python
a2b5 = streamline(r,2.,5.,0.)
x,y = cart(r,a2b5)
```

```python
plt.plot(x,y)
```

```python
a10b3 = streamline(r,10.,3.,0.)
x,y = cart(r,a10b3)
```

```python
plt.plot(x,y)
```

```python
from scipy.integrate import odeint
```

```python
def laneem_derivs(coords, xi, n):
    theta,u = coords
    if xi == 0:
        dtheta_dxi = 0
    else:
        dtheta_dxi = u/xi**2
    du_dxi = -xi**2*theta**n
    return [dtheta_dxi, du_dxi] #time derivatives
```

```python
# set up the grid
xi = np.linspace(0,10,num=1000)
#initial conditions
coords0 = [1.,0.] 
```

```python
#first analytic solution
n0 = 0.
sol_0 = odeint(laneem_derivs, coords0, xi, args=(n0,))

#second analytic solution
n1 = 1.
sol_1 = odeint(laneem_derivs, coords0, xi, args=(n1,))

#third analytic solution
n5 = 5.
sol_5 = odeint(laneem_derivs, coords0, xi, args=(n5,))
```

```python
plt.plot(xi, sol_0[:,0], 'violet', label = 'n=0')
plt.plot(xi, sol_1[:,0], 'purple', label = 'n=1')
plt.plot(xi, sol_5[:,0], 'teal', label = 'n=5')
plt.xlabel(r'$\xi$')
plt.ylabel('Density')
plt.legend()
plt.ylim([0,1])
#where soln_0 is parabolic
```

```python
#n = 1.5
n15 = 1.5
sol_15 = odeint(laneem_derivs, coords0, xi, args=(n15,))

#n = 3
n3 = 3.
sol_3 = odeint(laneem_derivs, coords0, xi, args=(n3,))
```

```python
plt.plot(xi, sol_0[:,0], 'violet', label = 'n=0')
plt.plot(xi, sol_1[:,0], 'purple', label = 'n=1')
plt.plot(xi, sol_15[:,0], 'cyan', label = 'n=1.5')
plt.plot(xi, sol_3[:,0], 'teal', label = 'n=3')
plt.plot(xi, sol_5[:,0], 'blue', label = 'n=5')
plt.xlabel(r'$\xi$')
plt.ylabel('Density')
plt.legend()
plt.ylim([0,1])
#where soln_0 is parabolic
```
